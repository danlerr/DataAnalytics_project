# EDA

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = pd.read_csv('./data/data.csv', delimiter=',')
#data.describe()
data
```

```python
# Divisione feature e target
X = data.drop(columns='Bankrupt?')
y = data['Bankrupt?']
```

```python
pca_model = PCA(n_components=2)
pca_coordinates = pca_model.fit_transform(data)
plt.scatter(pca_coordinates[:, 0], pca_coordinates[:, 1], c=y, cmap=plt.cm.get_cmap('inferno',6))
plt.colorbar(ticks=range(6), label='digit value')
PVE = pca_model.explained_variance_ratio_
PVE_cum = np.sum(PVE)
print(f'PVE:{PVE}, cumulative PVE:{PVE_cum}')
```
